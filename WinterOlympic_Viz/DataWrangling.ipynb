data sourced from https://www.kaggle.com/the-guardian/olympic-games/version/1?select=winter.csv

```python
import pandas as pd
```

```python
df = pd.read_csv("winter.csv")
```

```python
df.head()
```

```python
df = df[["Year","Sport","Country","Medal"]]
```

```python
df.head()
```

```python
df.groupby(["Sport", "Country"])["Medal"].count().reset_index(name="Medal_Count")
```

```python
df.rename(columns = {"Country":"Code"},inplace = True)
df.head()
```

```python
df.to_csv("Country-Sport-Medals.csv")
```

```python
df_dict = pd.read_csv("dictionary.csv")
```

```python
df_dict.head()
```

```python
df_country = df.set_index('Code').join(df_dict.set_index('Code'))
df_country.head()
```

```python
df_final = df_country[["Sport","Country","Medal"]]
df_final.head()
```

```python
df_medals = df_final.groupby(by = ["Sport","Country"]).count()
df_medals.head(55)
```

```python
df_medals.replace(to_replace = "United States", value = "USA", inplace = True)
df_medals.head(55)
```

```python
df_medals.to_csv("country_medal_count.csv")
```
